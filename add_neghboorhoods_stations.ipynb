```python
def get_neigh(lat, lon):
    reverse_geocode_result = gmaps.reverse_geocode((lat, lon))
    
    for geocode in reverse_geocode_result:
        dct = geocode['address_components']
        for s in dct:
            if 'neighborhood' in s['types']:
                return s['short_name']
    'locality', u'political'
    for geocode in reverse_geocode_result:
        dct = geocode['address_components']
        for s in dct:
            if ('locality' in s['types']) and ('political' in s['types']):
                return s['short_name']
    return None

def get_neigh_by_res(reverse_geocode_result):    
    for geocode in reverse_geocode_result:
        dct = geocode['address_components']
        for s in dct:
            if 'neighborhood' in s['types']:
                return s['short_name']
    'locality', u'political'
    for geocode in reverse_geocode_result:
        dct = geocode['address_components']
        for s in dct:
            if ('locality' in s['types']) and ('political' in s['types']):
                return s['short_name']
    return None
```

```python
df_stations['neigh'] = ""
df_stations
```

```python
from tqdm import  tqdm_notebook
quarts = ['Southeast Washington', 
          'Northeast Washington',
          'Northwest Washington', 
          'Southwest Washington']
nbr_list = []
for i in tqdm_notebook(range(df_stations.shape[0])):
    
    lat = df_stations.iloc[i].lat
    lon = df_stations.iloc[i].lon
    reverse_geocode_result = gmaps.reverse_geocode((lat, lon))
    nbr_str = get_neigh_by_res(reverse_geocode_result)
    
    if nbr_str in quarts:
        reverse_geocode_result.remove(reverse_geocode_result[0])
        df_stations.neigh.iloc[i] = get_neigh_by_res(reverse_geocode_result)
    else:
        nbr_list.append(nbr_str)
        df_stations.neigh.iloc[i] = nbr_str
        
df_stations
```
