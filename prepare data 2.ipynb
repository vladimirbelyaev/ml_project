```python
def plot_with_T(values):
    values = values.copy().values
    values[values == 'T'] = -1.0
    values = values.astype(np.float)
    plt.plot(np.arange(df_weather.index.values.shape[0]), values)
    plt.plot(np.arange(df_weather.index.values.shape[0]), pd.to_datetime(df_weather.index).year - 2014)
```
