```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,5)
```

```python
data = pd.read_csv('trips_with_weather.csv')
data.head()
```

```python
data.drop(data.columns[0], axis = 1, inplace=True)
data.drop(['Station Number', 'short_name'], axis = 1, inplace=True)
data.head()
```

```python
data = data.join( pd.get_dummies(data.day_part, prefix='day_part_'))

# data = data.join(pd.get_dummies(data.station_id, prefix='st_id_'))

data.drop('day_part', axis = 1, inplace=True)

data
```

```python
train_data = data[pd.to_datetime(data.date).dt.year != 2017].copy()
train_data
```

```python
test_data = data[pd.to_datetime(data.date).dt.year == 2017].copy()
test_data
```

```python
df_test = test_data.copy()
df_test.drop(['date', 'capacity'], axis = 1, inplace=True)
df_test.head()
```

```python
df_train = train_data.copy()
df_train.drop(['date', 'capacity'], axis = 1, inplace=True)
df_train.head()
```

```python
# словарь номер станции - данные по ней
train_stations_data = {}

for st_id in np.sort(df_train.station_id.unique()):
    train_stations_data[st_id] = df_train[df_train.station_id == st_id]
```

```python
# осредняем по каждой станции Input Output
train_mean_stations = {}

for st_id in train_stations_data:
    mean_input = train_stations_data[st_id].Input.mean()
    mean_output = train_stations_data[st_id].Output.mean()
    train_mean_stations[st_id] = [mean_input, mean_output]
train_mean_stations[1][0]
```

```python
# заполняем предсказание
```

```python
X_train = df_train.drop(['Input', 'Output'], axis = 1).values
y_train = df_train[['Input', 'Output']].values
```

```python
X_test = df_test.drop(['Input', 'Output'], axis = 1).values
y_test = df_test[['Input', 'Output']].values
```

```python
# в данных за 2015 - 2016 года и за 2017 есть различия
np.setdiff1d(df_test.station_id.unique(), df_train.station_id.unique())
```

```python
# заполняем предсказание средним по станции
df_result = pd.DataFrame(df_test.station_id, columns=['station_id', 'Input', 'Output'])
for st_id in train_mean_stations:
    df_result.Input[df_result.station_id == st_id] = train_mean_stations[st_id][0]
    df_result.Output[df_result.station_id == st_id] = train_mean_stations[st_id][1]
    
df_result.Input.fillna(df_result.Input.mean() , inplace = True)
df_result.Output.fillna(df_result.Output.mean(), inplace = True)
df_result
```

```python
my_y = df_result[['Input', 'Output']].values
my_y
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, my_y)
```
